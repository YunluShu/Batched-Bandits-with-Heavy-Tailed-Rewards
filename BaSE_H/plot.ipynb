```python
import os

import matplotlib.image as mpimg
import matplotlib.pyplot as plt

# Notebook 环境没有 __file__；用当前工作目录作为读取/保存路径
root = os.getcwd()

left_path = os.path.join(root, "plot_T.png")
right_path = os.path.join(root, "plot_TM.png")
out_path = os.path.join(root, "plot_merged.png")

left = mpimg.imread(left_path)
right = mpimg.imread(right_path)

fig, axs = plt.subplots(1, 2, figsize=(12, 5))
axs[0].imshow(left)
axs[0].axis("off")
axs[1].imshow(right)
axs[1].axis("off")

plt.subplots_adjust(
    wspace=0.02,
    hspace=0.0,
    left=0.0,
    right=1.0,
    top=1.0,
    bottom=0.0,
)

fig.savefig(out_path, dpi=300, bbox_inches="tight", pad_inches=0)
print(out_path)
```
